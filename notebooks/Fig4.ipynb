```python
from pylab import *
import os

matplotlib.rc('font', size=34)

home = os.getcwd() + "/../2Daxi"
repo2D_L1 = home + "/2Daxi_L1_mesh2_dt_0.0003_phi_0.00/"
repo2D_L5 = home + "/2Daxi_L5_mesh2_dt_0.0003_phi_0.00/"
repo2D_L10 = home + "/2Daxi_L10_mesh2_dt_0.0003_phi_0.00/"
repo2D_L50 = home + "/2Daxi_L50_mesh2_dt_0.0003_phi_0.00/"
repo2D_L100 = home + "/2Daxi_L100_mesh2_dt_0.0003_phi_0.00/"
repo2D_L200 = home + "/2Daxi_L200_mesh2_dt_0.0003_phi_0.00/"

# Velocity
vel_L1_dt6 = loadtxt(repo2D_L1 + "vavg.dat").transpose()
vel_L5_dt6 = loadtxt(repo2D_L5 + "vavg.dat").transpose()
vel_L10_dt6 = loadtxt(repo2D_L10 + "vavg.dat").transpose()
vel_L50_dt6 = loadtxt(repo2D_L50 + "vavg.dat").transpose()
vel_L100_dt6 = loadtxt(repo2D_L100 + "vavg.dat").transpose()
vel_L200_dt6 = loadtxt(repo2D_L200 + "vavg.dat").transpose()

tv_L1_dt6, v_L1_dt6 = vel_L1_dt6[:, 1050:1380]
tv_L5_dt6, v_L5_dt6 = vel_L5_dt6[:, 1050:1380]
tv_L10_dt6, v_L10_dt6 = vel_L10_dt6[:, 1050:1380]
tv_L50_dt6, v_L50_dt6 = vel_L50_dt6[:, 1050:1380]
tv_L100_dt6, v_L100_dt6 = vel_L100_dt6[:, 1050:1380]
tv_L200_dt6, v_L200_dt6 = vel_L200_dt6[:, 1050:1380]

# Position
pos_L1_dt6 = loadtxt(repo2D_L1 + "position.dat").transpose()
pos_L5_dt6 = loadtxt(repo2D_L5 + "position.dat").transpose()
pos_L10_dt6 = loadtxt(repo2D_L10 + "position.dat").transpose()
pos_L50_dt6 = loadtxt(repo2D_L50 + "position.dat").transpose()
pos_L100_dt6 = loadtxt(repo2D_L100 + "position.dat").transpose()
pos_L200_dt6 = loadtxt(repo2D_L200 + "position.dat").transpose()

t_L1_dt6, p_L1_dt6 = pos_L1_dt6
t_L5_dt6, p_L5_dt6 = pos_L5_dt6
t_L10_dt6, p_L10_dt6 = pos_L10_dt6
t_L50_dt6, p_L50_dt6 = pos_L50_dt6
t_L100_dt6, p_L100_dt6 = pos_L100_dt6
t_L200_dt6, p_L200_dt6 = pos_L200_dt6

# Peaks
peak_L1_dt6 = loadtxt(repo2D_L1 + "peaks.dat").transpose()
peak_L5_dt6 = loadtxt(repo2D_L5 + "peaks.dat").transpose()
peak_L10_dt6 = loadtxt(repo2D_L10 + "peaks.dat").transpose()
peak_L50_dt6 = loadtxt(repo2D_L50 + "peaks.dat").transpose()
peak_L100_dt6 = loadtxt(repo2D_L100 + "peaks.dat").transpose()
peak_L200_dt6 = loadtxt(repo2D_L200 + "peaks.dat").transpose()

# Showing last two peaks
pk_L1_dt6, tp_L1_dt6 = peak_L1_dt6[:,-2:]
pk_L5_dt6, tp_L5_dt6 = peak_L5_dt6[:,-2:]
pk_L10_dt6, tp_L10_dt6 = peak_L10_dt6[:,-2:]
pk_L50_dt6, tp_L50_dt6 = peak_L50_dt6[:,-2:]
pk_L100_dt6, tp_L100_dt6 = peak_L100_dt6[:,-2:]
pk_L200_dt6, tp_L200_dt6 = peak_L200_dt6[:,-2:]

microm = 1e3 # From mm to mu m

f, ms = subplots(1,2,figsize=(20,8))
subplots_adjust(wspace=0.4)

ms[0].plot(tv_L1_dt6,v_L1_dt6*microm, \
             tv_L5_dt6,v_L5_dt6*microm, \
             tv_L10_dt6,v_L10_dt6*microm, \
             tv_L50_dt6,v_L50_dt6*microm, \
             tv_L100_dt6,v_L100_dt6*microm, \
             tv_L200_dt6,v_L200_dt6*microm, \
             linewidth = 4)

ms[0].set_xlabel('Time (s)')
ms[0].set_ylabel('Velocity (µm\s)')
ms[0].locator_params(axis='x', nbins=2)
ms[0].axhline(y=0, color='k')


l1, l2, l3, l4, l5, l6 = ms[1].plot(t_L1_dt6,p_L1_dt6*microm, \
                                    t_L5_dt6,p_L5_dt6*microm, \
                                    t_L10_dt6,p_L10_dt6*microm, \
                                    t_L50_dt6,p_L50_dt6*microm, \
                                    t_L100_dt6,p_L100_dt6*microm, \
                                    t_L200_dt6,p_L200_dt6*microm, \
                                    linewidth = 4)

ms[1].set_prop_cycle(rcParams['axes.prop_cycle']) # Reset color cycle
ms[1].plot(tp_L1_dt6,pk_L1_dt6*microm, \
             tp_L5_dt6,pk_L5_dt6*microm, \
             tp_L10_dt6,pk_L10_dt6*microm, \
             tp_L50_dt6,pk_L50_dt6*microm, \
             tp_L100_dt6,pk_L100_dt6*microm, \
             tp_L200_dt6,pk_L200_dt6*microm, \
             marker='s', \
             markersize=10, \
             linestyle=':')

# Computing slopes from 2 last peaks
slope_L1 = (peak_L1_dt6[0,-1]*microm - peak_L1_dt6[0,-2]*microm)/(peak_L1_dt6[1,-1] - peak_L1_dt6[1,-2])
slope_L5 = (peak_L5_dt6[0,-1]*microm - peak_L5_dt6[0,-2]*microm)/(peak_L5_dt6[1,-1] - peak_L5_dt6[1,-2])
slope_L10 = (peak_L10_dt6[0,-1]*microm - peak_L10_dt6[0,-2]*microm)/(peak_L10_dt6[1,-1] - peak_L10_dt6[1,-2])
slope_L50 = (peak_L50_dt6[0,-1]*microm - peak_L50_dt6[0,-2]*microm)/(peak_L50_dt6[1,-1] - peak_L50_dt6[1,-2])
slope_L100 = (peak_L100_dt6[0,-1]*microm - peak_L100_dt6[0,-2]*microm)/(peak_L100_dt6[1,-1] - peak_L100_dt6[1,-2])
slope_L200 = (peak_L200_dt6[0,-1]*microm - peak_L200_dt6[0,-2]*microm)/(peak_L200_dt6[1,-1] - peak_L200_dt6[1,-2])

ms[1].set_xlabel('Time (s)')
ms[1].set_ylabel('Position (µm)')
ms[1].axhline(y=0, color='k')

# Common legend to both plots
lgd = f.legend((l1, l2, l3, l4, l5, l6), \
               ('1 mm (%.1f $\mu$m/s)'%(-slope_L1), \
                '5 mm (%.1f $\mu$m/s)'%(-slope_L5), \
                '10 mm (%.1f $\mu$m/s)'%(-slope_L10), \
                '50 mm (%.1f $\mu$m/s)'%(-slope_L50), \
                '100 mm (%.1f $\mu$m/s)'%(-slope_L100), \
                '200 mm (%.1f $\mu$m/s)'%(-slope_L200)), \
             loc = "upper center", prop={'size':28}, ncol=2, \
             bbox_to_anchor=(0.5, 1.2), labelspacing = 0.2, columnspacing=10) 

savefig('Fig4.png', bbox_extra_artists=(lgd,), bbox_inches='tight')
show()
```
