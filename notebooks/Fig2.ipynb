```python
from pylab import *
import os

home = os.getcwd() + "/../3D"
pulsatile_dir = home + "/C0075_fine_results_f_10_gradp_0.0000_phi_0.00/"
static_dir = home + "/C0075_fine_results_f_10_gradp_0.1995_phi_0.00/"
systemic_dir = home + "/C0075_fine_results_f_10_gradp_0.0000_phi_0.10/"
rigid_dir = home + "/C0075_fine_results_f_10_gradp_0.0000_phi_0.00_rigid/"

matplotlib.rc('font', size=34)
pulsatile_pos = loadtxt(pulsatile_dir + 'position.dat').transpose()
pulsatile_vel = loadtxt(pulsatile_dir + 'vavg.dat').transpose()

static_pos = loadtxt(static_dir + 'position.dat').transpose()
static_vel = loadtxt(static_dir + 'vavg.dat').transpose()

systemic_pos = loadtxt(systemic_dir + 'position.dat').transpose()
systemic_vel = loadtxt(systemic_dir + 'vavg.dat').transpose()

rigid_pos = loadtxt(rigid_dir + 'position.dat').transpose()
rigid_vel = loadtxt(rigid_dir + 'vavg.dat').transpose()

t,d = pulsatile_pos
t2, v = pulsatile_vel[:,1:]
ts,ds = static_pos
ts2, vs = static_vel[:,1:]
tb, db = systemic_pos
tb2, vb = systemic_vel[:,1:]
tr, dr = rigid_pos
tr2, vr = rigid_vel[:,1:]

f, (ax1, ax2, ax0) = subplots(1,3,figsize=(32,8))
subplots_adjust(wspace=0.4)

ax0.plot(t,1e3*d,ts, 1e3*ds, tb,1e3*db, tr, 1e3*dr, linewidth = 4)
ax0.legend(['Pulsatile', 'Static','Systemic', 'Rigid'],prop={'size':28})
ax0.set_xlabel('Time (s)')
ax0.set_ylabel('Position ($\mu$m)')
ax1.plot(t2,1e3*v,ts2, 1e3*vs, tb2, 1e3*vb, tr2, 1e3*vr, linewidth = 4)
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Velocity ($\mu$m/s)')
ax2.plot(t2,1e3*v,ts2, 1e3*vs, tb2, 1e3*vb, tr2, 1e3*vr, linewidth = 4)
ax2.set_xlim([0.325,0.425])
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Velocity ($\mu$m/s)')
ax0.axhline(y=0, color='k')
ax1.axhline(y=0, color='k')
ax2.axhline(y=0, color='k')
savefig('Fig2.png')
show()
```
